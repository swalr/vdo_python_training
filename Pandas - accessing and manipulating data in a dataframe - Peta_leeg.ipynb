# Loc en iloc

Twee hele nuttige functies in pandas zijn loc en iloc. Die functies gebruikt je vaak om specifieke kolommen, rijen en cellen te selecteren. 

Loc wordt gebruikt om specifieke rijen en kolommen te selecteren op basis van de kolom- en rijnaam. Iloc wordt gebruikt om rijen en kolommen te selecteren op basis van hun positie in de dataframe.

Loc en iloc kunnen allebei worden gebruikt om nieuwe waarde aan cellen te geven.

Een groot verschil met wat we tot nu toe hebben gedaan is dat je met loc en iloc specifieke rijen en kolommen tegelijkertijd kan selecteren. Dat is wel handig wanneer je een dataset hebt met 300 kolommen en 20'000 rijen!

Wat is dan het verschil tussen loc en iloc? Loc gebruik je om rijen en kolommen te selecteren op basis van hun naam (e.g., loc['Type auto']), terwijl iloc wordt gebruikt om rijen en kolommen te selecteren op basis van hun positie in de dataframe (e.g., iloc[1]).

De syntax voor loc en iloc is:
-  loc[rij_naam, kolom_naam]
-  iloc[rij_positie, kolom_positie]
