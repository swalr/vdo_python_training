### (middag-)opdracht

In deze opdracht gaan we met een cleane dataset wat statistische analyses uitvoeren en daarna mooie plots maken.

Importeer eerst numpy en pandas, en daarna deze dataset als df3. Haal alle NaN"s eruit.

Vraag de statistieken van de dataframe op. Hoeveel rijen en kolommen bevat de dataset? Hoeveel en welke landen zijn er? Hoeveel jongens en hoeveel meisjes zitten er in deze dataset? 

Je kunt kruistabellen maken door pandas.crosstab te gebruiken. Als je bijvoorbeeld het land tegenover geslacht wil zetten, kan dat zo:

`pandas.crosstab(df3.CNT,df3.GENDER)`

Doe dit nu voor geslacht en de scores op de variabele testanxiety. Wat zie je nu verschijnen? Is dit een handige weergave?

Je kunt kerngegevens opvragen van een dataframe door `pd.describe` te gebruiken. Probeer dit nu voor de hele dataframe en daarna op een variabele naar keuze. Rond af naar 1 cijfer achter de komma.

Zoals je ziet, zijn alleen de descriptives van de numerieke variabelen meegenomen. Je kunt `include=[object]` toevoegen om categorische descriptives te krijgen.

### Plots

Voor plots gebruiken we een nieuwe module genaamd plotnine. Met plotnine kun je ggplots gebruiken, een mooie manier om je data te visualiseren. Importeer plotnine nu:

`import plotnine`

De manier van plotten in plotnine gaat via ggplot, misschien jullie wel bekend vanuit R. [Hier](https://monashdatafluency.github.io/python-workshop-base/modules/plotting_with_ggplot/) is een mooie tutorial te vinden van ggplot, lees deze door en kijk of je het begrijpt.

Probeer een histogram te maken van leeftijden. Speel met de verschillende functies van ggplot en kijk wat je allemaal kan maken.
